The decision was made to drop the rows containing NA values for bedrooms and bathrooms. If the average was taken of each column individually and filled the NAs cells, the count for houses with those beds/baths will be greatly skewed. It will in turn affect the price on the house, possibly increasing it more than its worth or lowering it to less than what its worth. 

```python
## Dropping the empty rows for bedrooms and bathrooms column
df= df.dropna(subset = ('bedrooms','bathrooms'))
rows_new = df.shape[0] # storing number of rows after dropped NANs

## Percent of data dropped based on rows
print(f"The percent of data removed is {(rows_old-rows_new)/rows_old * 100} %") #about 10%


## Checking how many rows are empty now
df.isna().sum()
```

Removing all the rows with NAs for bedroom and bathroom did create a considerable dent in the dataframe, but there is still 90% of the data left to analyze. Let's also look and see if there's any outliers that need to be removed.

```python
## Looking at the summary statistics for bedroom and bathroom
display(df['bedrooms'].describe())
display(df['bathrooms'].describe())

## Looking at the unique values for bedrooms and bathrooms
display(df['bedrooms'].unique())
display(df['bathrooms'].unique())
## getting the outliers for bedroom and bathroom
bed_min, bed_max = df['bedrooms'].min(), df['bedrooms'].max()
bath_min, bath_max = df['bathrooms'].min(), df['bathrooms'].max()
print(f" The maximum and minimum number of bedrooms are: {bed_max,bed_min}")
print(f" The maximum and minimum number of bathrooms are: {bath_max, bath_min}")
```

The rows containing 0 or 33 bedrooms will be dropped for bedrooms and the rows containing 0 for bathrooms will be dropped. Usually people will look for houses with at least 1 bedroom and part of a bathroom, which is why the zeroes will be dropped and the outlier of 33 bedrooms will be dropped. 

```python
## Removing rows with 0 for bathroom and bedroom and removing rows with 33 for bedrooms
## .index selects the rows that passes the conditions
## inplace = True makes changes to the existing dataframe
df.drop(df[(df['bedrooms'] == 0) | (df['bedrooms'] == 33) | (df['bathrooms'] ==  0)].index, inplace=True)
new_rows = df.shape[0]

## Percent of data dropped based on rows
print(f"The percent of data removed is {(rows_old-rows_new)/rows_old * 100} %") #about 10%
```

There's still 90% of the data left even with removing the outliers.

To fill in the NANs for sqft_living and sqft_lot, the data was first grouped by the number of bedrooms and bathrooms to determine all the combinations between the two.

```python
## Checking the different combinations for each bedroom and bathroom count
bed_bath_combo = df.groupby(['bedrooms', 'bathrooms']).size().reset_index().rename(columns = {0:'count'})
bed_bath_combo
```

There are 129 combinations between bedrooms and bathrooms. From here, another grouping is performed on the data to get the average of sqft_living and sqft_lot based on those combinations. The first table displayed will be for sqft_living and the second table will be for sqft_lot.

```python
## Getting the sqft_living average of all the different combinations of beds/baths
avg_sqft_living = df.groupby(['bedrooms', 'bathrooms'])['sqft_living'].mean().reset_index(name = 'average')
display(avg_sqft_living)
display(avg_sqft_living.isna().sum())

## Getting the sqft_lot average of all the different combinations of beds/baths
avg_sqft_lot = df.groupby(['bedrooms', 'bathrooms'])['sqft_lot'].mean().reset_index(name = 'average')
display(avg_sqft_lot)
avg_sqft_lot.isna().sum() #checking for NANs in the dataframe
```

In the table for sqft_living, there is still one row with NA. That NA will be filled in later with the overall average of sqft_living. To see which combination of bed/bath it was, we can subset the count table earlier.

```python
living_na = avg_sqft_living['average'].isna()
living_na = living_na[living_na == True].index
bed_bath_combo.iloc[living_na] #looking in the bed_bath_combo dataframe, there was only one instance of this combo

## Removing the row of NA in the avg_sqft_living dataframe, will fill that in with an overall avg later
avg_sqft_living = avg_sqft_living.dropna(subset = 'average')
```

It was the combination of 6 bathrooms and 7.75 bathrooms that had an NA value. But since there is no other instance of this combination, this row will wait until the end to be filled in with the overall average. We'll also drop this row from the avg_sqft_living table in order to process the rest of the NA values for df.

```python
## This for loop takes the combination of the beds/bath and gets the average for that combo.
## Next, it finds which indexes have that combination for sqft_living.
## Then, it gets the indexes within that index that have NA values and replaces the NA value for the average.

for x, y in zip(avg_sqft_living['bedrooms'],avg_sqft_living['bathrooms']):
       new_val = avg_sqft_living[(avg_sqft_living['bedrooms'] == x) & (avg_sqft_living['bathrooms'] == y)]['average'].values
       new_val = new_val[0]
       row_indx = df[(df['bedrooms'] == x) & (df['bathrooms'] == y)].index
       new_indx = df['sqft_living'][row_indx].isna()
       na_indx = new_indx[new_indx == True].index
       df.loc[na_indx,'sqft_living'] = new_val

df.isna().sum()
```

You can see that all the NAs have been accounted for except for 1. That will be filled in after running the same loop for sqft_lot.

```python
## This for loop takes the combination of the beds/bath and gets the average for that combo.
## Next, it finds which indexes have that combination for sqft_living.
## Then, it gets the indexes within that index that have NA values and replaces the NA value for the average.

for x, y in zip(avg_sqft_lot['bedrooms'], avg_sqft_lot['bathrooms']):
       new_val = avg_sqft_lot[(avg_sqft_lot['bedrooms'] == x) & (avg_sqft_lot['bathrooms'] == y)]['average'].values
       new_val = new_val[0]
       row_indx = df[(df['bedrooms'] == x) & (df['bathrooms'] == y)].index
       new_indx = df['sqft_lot'][row_indx].isna()
       na_indx = new_indx[new_indx == True].index
       df.loc[na_indx,'sqft_lot'] = new_val
    
df.isna().sum()

## Filling in the last datapoint with the overall average of sqft_living
df['sqft_living'].fillna(df['sqft_living'].mean(),inplace = True)

df.isna().sum()
```

No more NA values in the dataset.
